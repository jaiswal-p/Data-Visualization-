## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
new_df = pymoli_df.loc[:, ['Item ID', 'Item Name', 'Price']]
# create object and by grouping the item data frame by "Item ID" and "item name"
groupby_df = new_df.groupby(["Item ID", "Item Name"])
purchase_count = groupby_df["Price"].count()   
purchase_value = groupby_df["Price"].sum() 
purchase_value = purchase_value.round(2)

item_price =purchase_value/purchase_count  
item_price=item_price.round(2)

#Dataframe to display the value above 
summery_item=pd.DataFrame({"Purchase Count":purchase_count,"Item Price":"$"+item_price.apply(str),"Total Purchase Value": "$" + purchase_value.apply(str)})
# sort the data frame by purchase count 
summery_item.sort_values("Purchase Count",ascending=False).head(5)

                         
```
